### Merge all in a one file

```python
mergeData = pd.DataFrame()

for file in allMonths:
    single_file = pd.read_csv("./Desktop/DA 1/Sales_Data/"+file)
    mergeData = pd.concat([mergeData , single_file])
    
mergeData.to_csv("./Desktop/DA 1/Sales_Data/merge.csv", index=False)
```

```python
aData = pd.read_csv('./Desktop/DA 1/Sales_Data/merge.csv')
```

```python
aData.head()
```

## Data Arranging

#### Adding a month column


```python
aData["Month"] = aData['Order Date'].str[0:2]
```

```python
aData
```

```python
aData['Month'] = aData['Month'].str.replace(r'/$', '', regex=True)
```

```python
aData
```

#### Clean the data


```python
aData = aData.dropna(how = 'all')
```

```python
aData
```

```python
aData = aData[aData['Order Date'].str[0:2] != 'Or']
```

```python
aData
```

```python
aData['Month'] = aData['Month'].astype('int32')
```

```python
aData.head(20000)
```

#### Adding a Sales column

```python
aData["Sales"] = 2
aData
```

```python
aData['Price Each'] = pd.to_numeric(aData['Price Each'])
aData['Quantity Ordered'] = pd.to_numeric(aData['Quantity Ordered'])
aData["Sales"] = aData['Quantity Ordered']* aData['Price Each']
```

```python
aData
```


## Best Month of Sales

```python
ans = aData.groupby("Month").sum(numeric_only=True)
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(0,12)
```

```python
plt.bar(months,ans["Sales"])
plt.xlabel('Months')
plt.ylabel('Sales in USD')
plt.title('Monthly Total Sales in USD')
plt.xticks(rotation=45)
plt.show()
```

## Sales by City

#### Extract city

```python
aData['City'] = aData['Purchase Address'].str.split(',').str[1].str.strip() + " " + aData['Purchase Address'].str.split(',').str[2].str.strip()
```

```python
aData['City']= aData['City'].str.slice(stop=-5)
```

```python
aData.tail(75000)
```

```python
ans2 = aData.groupby('City').sum(numeric_only = True)
ans2
```

```python
c = aData['City'].unique()
plt.bar(c,ans2["Sales"])

plt.xlabel('City')
plt.ylabel('Sales in USD')
plt.title(' Total Sales by Cities')
plt.xticks(rotation=45)
plt.show()
```
